**Exploração das Variáveis qualitativas nominais**

Sexo


```python
dados['SEXO'].unique() 
```

```python
dados['SEXO'].value_counts()
```

> Temos no DF de março de 2022 a quantidade de 10064 pessoas em situacao de rua sendo que 8998 são do sexo masculino e 1066 do sexo feminino.

```python
dados['SEXO'].value_counts(normalize = True) * 100
```

```python
dados_sex_perc = dados['SEXO'].value_counts(normalize = True) * 100
```

```python
dados_sex = dados['SEXO'].value_counts()
```

```python
dados_sex.plot(kind='pie', autopct='%1.2f%%') 

```

> Temos dentro desse DF o quantitativo de 89,41% de homens e 10,59% de mulheres vivendo em situação de rua em Belo Horionte cadastradas no CADUNICO da PBH

```python
dist_freq_sexo = pd.DataFrame({'Frequência': dados_sex, 'Porcentagem (%)': dados_sex_perc})
dist_freq_sexo.rename_axis('Sexo', axis = 'columns', inplace = True)
dist_freq_sexo
```

Raça

```python
dados['COR_RACA'].unique() 
```

```python
dados['COR_RACA'].value_counts()
```

Sexo X Raça 

```python
frequencia_sex_cor = pd.crosstab(dados['SEXO'],
                         dados['COR_RACA'])
frequencia_sex_cor
```

```python
frequencia_sex_cor_perc =  pd.crosstab(dados['SEXO'],
                         dados['COR_RACA'],
                         normalize =True) * 100
frequencia_sex_cor_perc
```

```python
frequencia_sex_cor_perc =  pd.crosstab(dados['SEXO'],
                         dados['COR_RACA'],
                         aggfunc = 'mean',
                         values = dados['IDADE']) 
frequencia_sex_cor_perc
```

ANALISE DO GRUPO

```python
dados_masc = dados.query('SEXO == "MASCULINO"')
dados_fem = dados.query('SEXO == "FEMININO"')
```

```python
dados_masc['IDADE'].mean().round(0)
print('A idade média da populaçao em situacao de rua é de aproximadamente {} anos para pessoas autodeclaradas do sexo masculino em 2022'.format(dados_masc['IDADE'].mean().round(0)))
```

```python
dados_fem['IDADE'].mean().round(0)
print('A idade média da populaçao em situacao de rua é de aproximadamente {} anos para pessoas autodeclaradas do sexo feminino em 2022'.format(dados_fem['IDADE'].mean().round(0)))
```
